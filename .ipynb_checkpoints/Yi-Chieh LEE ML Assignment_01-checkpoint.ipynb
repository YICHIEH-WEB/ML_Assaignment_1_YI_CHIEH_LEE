    b)	(1 point) With a for loop, draw and study the values in each of the remaining W columns. (hint for each w column draw histogram and box plot)

```python
# Your code

import matplotlib.pyplot as plt

 # axis check the it vertically which is Week and drop the columns contains ALL NAN
df = bb_df.dropna(axis=1, how='all')

 #w_cols is a Series like W1,W2,W3 which contains all rows in the column
w_cols = [c for c in df.columns if c.startswith('w')] 

# using for loop for draw the graphics 
for col in w_cols:
    data = df[col].dropna() # iterate each items within the Series, drop the nan.
    
    # histogram
    plt.figure()
    plt.hist(data, bins=10)
    plt.title(f'Histogram of {col}')
    plt.xlabel('Rank')
    plt.ylabel('Frequency')
    plt.show()
    
    # boxplot
    plt.figure()
    plt.boxplot(data)
    plt.title(f'Box Plot of {col}')
    plt.ylabel('Rank')
    plt.show()

```

6. The director of graduate admission at a large university is analyzing the relationship between scores on the math portion of Graduate Record Examination (GRE) and subsequent performace in graduate school, as measured by student`s avrage (GPA). the accompaanying data file shows GPA and GRE of 24 students.

```python
import numpy as np
import pandas as pd
gpaDf =  pd.read_excel("https://www.dropbox.com/scl/fi/24jfn2u4rsn6e45bvzkrs/GPA.xlsx?rlkey=9ecqtc90bmntcsucdy8gzyt24&dl=1")
gpaDf
```

    a) (2points) Find regression model equation for the model: GPA = b0 + B1 GRE

```python
import pandas as pd
import statsmodels.api as sm

# independent , dependent
X = sm.add_constant(gpaDf[["GRE"]])
y = gpaDf["GPA"]

# OLS  Train linear regression model
model = sm.OLS(y, X).fit()

print(model.summary())



# We can also see the coefficients of constant, intercept is (b0= 0.425) and coeffiecient of GRE is ( b1= 0.0041)
# Answer is in another chunk
```

```python
# another way to show b0, b1 and model equation
b0, b1 = model.params["const"], model.params["GRE"]

print(f"GPA = {b0:.4f}+{b1:.4f} * GRE")
```

    b) (1 point) what is a student`s prediction GPA if he/she scored 710 on math portion of GRE

```python
# Your answer


# Follwing on the last question. The equation is: GPA = 0.4256 + 0.0041 * GRE

def GPA_func(GRE):
    GPA = b0 + b1 * GRE
    return GPA
    
GPA = GPA_func(710)
print(f"The prediction GPA is {GPA}")
```

##### Excercise 7.1

10. The maintenance manager at a trucking company wants to build a regression model to forcast the time unyil the first engine overhaul(Time in years) based on four predictor variables for 25 trucks:
    1. anual miles driven (Miles in 1000s)
    2. avrage load weight (Load in tons)
    3. average driving speed (Speed in mph)
    4. oil change interval (Oil in 1000s of miles).

```python
import numpy as np
import pandas as pd
engineDf =  pd.read_excel("https://www.dropbox.com/scl/fi/dcoz9yw3f8yywtzy82f4z/Engine.xlsx?rlkey=n53hfjjsrddywktksj156jra5&dl=1")
engineDf.head()
```

    a) (2 point)find model regression equation (use all four predictor variables)

```python
# your answer
# https://www.statsmodels.org/stable/index.html
import statsmodels.api as sm

#independent, dependent 
X=engineDf[['Miles','Load','Speed','Oil']]
y=engineDf['Time']

# add intercept
X=sm.add_constant(X)

# build model using OLS linear regression
model=sm.OLS(y,X).fit()

model.summary()
```

```python
params = model.params # Only retrieve the coefficient
params
```

```python
print(f" Regression equation: Time = {params['const']:.4f} + {params["Miles"]:.4f} * Miles  + {params["Load"]:.4f} * Load + {params["Speed"]:.4f} * Speed + {params["Oil"]:.4f}* Oil")
```

    b) (1point) Predict the time before the first engine overhaul for particular truck driven 60000 miles per year with an average load of 22 tons, an average driving speed of 57 mph, and 18000 miles between oil changes.

```python
# Your answer

# create new data frame
# 60000 miles -> 60 (thousand)  # 18000 miles -> 18 (thousand)
new_data=pd.DataFrame({'constant':[1],"Miles":[60],"Load":[22],"Speed":[57],"Oil":[18]})
new_data
```

```python

# predict model
predicted_time = model.predict(new_data)
print(f'Predicted time before first engine overhaul: {predicted_time.iloc[0]:.2f} years')
```
